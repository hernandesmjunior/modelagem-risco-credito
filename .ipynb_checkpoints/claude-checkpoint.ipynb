<p style="font-size:16px;line-height:1.6;color:#2D2A32;">
O objetivo é estimar a <b>probabilidade de inadimplência</b> (default) de solicitantes de crédito
a partir do benchmark público <b>German Credit</b>. O problema é de classificação binária, mas o
que separa um exercício de Kaggle de um trabalho de risco de crédito não é o algoritmo: é a
<b>higiene de validação</b>, a <b>interpretabilidade</b> exigida por regulação e o <b>custo
assimétrico</b> do erro. Aprovar um mau pagador custa o principal emprestado; reprovar um bom
pagador custa uma margem. Os dois erros não valem o mesmo, e o modelo precisa refletir isso.</p>

<p style="font-size:15px;line-height:1.6;color:#2D2A32;">
O <b>German Credit</b> reúne 1.000 solicitações de crédito descritas por atributos financeiros e
demográficos, cada uma rotulada como bom (<i>good</i>) ou mau (<i>bad</i>) pagador. A taxa de maus
é de cerca de 30%, o que torna a classe de interesse a <b>minoritária</b> — situação típica em
crédito e que exige cuidado tanto na modelagem quanto na leitura das métricas.</p>

<div style="border-left:5px solid #FB923C;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Convenção do alvo</b><br><span style="color:#2D2A32;">Codificamos <b>default = 1</b> (mau pagador) e <b>adimplente = 0</b>. A classe positiva é o evento que queremos capturar. Recall sobre a classe 1 é, portanto, a fração de inadimplentes que o modelo efetivamente pega — a métrica que o negócio sente no bolso.</span></div>

<p style="font-size:15px;line-height:1.6;color:#2D2A32;">
Um scorecard de crédito não pode ser uma caixa-preta. Reguladores e áreas de risco exigem
<b>explicabilidade</b>: por que um cliente foi reprovado. Por isso o baseline aqui é uma
<b>regressão logística</b> — coeficientes lêem como odds — e o XGBoost entra como <b>challenger</b>
para medir quanto de sinal não-linear está sendo deixado na mesa.</p>

<div style="border-left:5px solid #7C3AED;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Decisão de tratamento</b><br><span style="color:#2D2A32;">Em <b>Saving accounts</b> e <b>Checking account</b>, o ausente não é ruído aleatório: significa 'cliente não possui esse produto'. Tratar como categoria própria ('no_account') preserva informação, e veremos que a própria ausência é preditiva. Imputar pela moda apagaria sinal. A imputação será feita <b>após</b> o split, aprendendo só do treino.</span></div>

<div style="background:linear-gradient(90deg,#7C3AED 0%,#A66CFF 100%);border-radius:10px;padding:14px 22px;margin:26px 0 10px 0;"><span style="color:#fff;font-size:26px;font-weight:900;letter-spacing:.5px;"><span style="opacity:.6;">6.</span>&nbsp;&nbsp;PARTIÇÃO TREINO / TESTE</span></div>

<div style="border-left:5px solid #EA580C;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Por que o split vem antes de tudo</b><br><span style="color:#2D2A32;">Qualquer estatística usada para transformar os dados — imputação, WOE, escalonamento — precisa ser aprendida <b>somente no treino</b> e aplicada ao teste. Calcular WOE ou imputar usando o dataset inteiro é vazamento (data leakage) e infla as métricas. Por isso particionamos agora, de forma estratificada para preservar os 30% de default nos dois conjuntos.</span></div>

```python
cat_cols = ["Sex", "Job", "Housing", "Saving accounts", "Checking account", "Purpose"]
num_cols = ["Age", "Credit amount", "Duration"]

X = credit[cat_cols + num_cols].copy()
y = credit["default"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y)

# Ausente = "não possui o produto": categoria informativa, não imputação
for c in ["Saving accounts", "Checking account"]:
    X_train[c] = X_train[c].fillna("no_account")
    X_test[c]  = X_test[c].fillna("no_account")
for c in cat_cols:
    X_train[c] = X_train[c].astype(str)
    X_test[c]  = X_test[c].astype(str)

print(f"Treino: {X_train.shape[0]}  |  default {y_train.mean():.1%}")
print(f"Teste : {X_test.shape[0]}  |  default {y_test.mean():.1%}")
```

<div style="background:linear-gradient(90deg,#7C3AED 0%,#A66CFF 100%);border-radius:10px;padding:14px 22px;margin:26px 0 10px 0;"><span style="color:#fff;font-size:26px;font-weight:900;letter-spacing:.5px;"><span style="opacity:.6;">7.</span>&nbsp;&nbsp;WEIGHT OF EVIDENCE E INFORMATION VALUE</span></div>

<p style="color:#2D2A32;line-height:1.6;">
O <b>Weight of Evidence</b> mede, para cada categoria, o quanto ela pende para bons ou maus
pagadores, na escala do log das chances:</p>
<p style="text-align:center;font-size:17px;color:#7C3AED;">
WoE = ln( %bons / %maus )&nbsp;&nbsp;&nbsp;&nbsp;
IV = &Sigma; ( %bons − %maus ) &times; WoE</p>
<p style="color:#2D2A32;line-height:1.6;">
O <b>Information Value</b> agrega o WoE numa nota única de poder preditivo. A régua clássica da
indústria: IV &lt; 0,02 inútil; 0,02–0,1 fraco; 0,1–0,3 médio; 0,3–0,5 forte; &gt; 0,5 suspeito de
vazamento. Calculamos tudo sobre as <b>categorias cruas</b> do treino.</p>

```python
def woe_table(feature, target):
    """Tabela WoE/IV para uma variável categórica (categorias cruas)."""
    d = pd.DataFrame({"cat": feature.astype(str), "y": target.values})
    g = d.groupby("cat")["y"].agg(n="count", bad="sum")
    g["good"] = g["n"] - g["bad"]
    g["dist_good"] = g["good"] / g["good"].sum()
    g["dist_bad"]  = g["bad"]  / g["bad"].sum()
    g["bad_rate"]  = g["bad"] / g["n"]
    g["WoE"] = np.log((g["dist_good"] + 1e-6) / (g["dist_bad"] + 1e-6))
    g["IV_contrib"] = (g["dist_good"] - g["dist_bad"]) * g["WoE"]
    return g.sort_values("WoE"), g["IV_contrib"].sum()

def information_value(X, y, cats, nums, n_bins=5):
    ivs = {}
    for c in cats:
        _, iv = woe_table(X[c], y); ivs[c] = iv
    for c in nums:                       # numéricas: binning por quantil
        binned = pd.qcut(X[c], n_bins, duplicates="drop").astype(str)
        _, iv = woe_table(binned, y); ivs[c] = iv
    return pd.Series(ivs).sort_values(ascending=False)

iv_ranking = information_value(X_train, y_train, cat_cols, num_cols)
iv_ranking.round(4).to_frame("Information Value")
```

```python
def iv_band(v):
    if v < 0.02:  return "inútil"
    if v < 0.10:  return "fraco"
    if v < 0.30:  return "médio"
    if v < 0.50:  return "forte"
    return "suspeito (leakage?)"

fig, ax = plt.subplots(figsize=(9, 5))
colors = [ORANGE if v >= 0.10 else PURPLE_L for v in iv_ranking.values]
ax.barh(iv_ranking.index, iv_ranking.values, color=colors, edgecolor="white")
for i, (name, v) in enumerate(iv_ranking.items()):
    ax.text(v + .008, i, f"{v:.3f} · {iv_band(v)}", va="center", color=INK, fontsize=10)
ax.invert_yaxis(); ax.set_xlim(0, iv_ranking.max()*1.35)
ax.set_title("Information Value por variável — poder preditivo bruto")
ax.set_xlabel("IV"); sns.despine(); plt.tight_layout(); plt.show()
```

<div style="border-left:5px solid #7C3AED;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Leitura</b><br><span style="color:#2D2A32;"><b>Checking account</b> (IV ≈ 0,62) domina e beira o limiar de 'suspeito'. Aqui não é vazamento: é uma variável genuinamente forte, típica desse benchmark. <b>Saving accounts</b> e <b>Duration</b> ficam na faixa média. <b>Job</b> e <b>Sex</b> são quase inúteis (IV &lt; 0,04) — candidatos naturais a exclusão, e no caso de Sex, também uma variável sensível que um modelo de crédito responsável evita.</span></div>

```python
# WoE detalhado da variável mais forte
woe_detail, iv_detail = woe_table(X_train["Checking account"], y_train)
print(f"Checking account — IV = {iv_detail:.4f}\n")
woe_detail[["n", "bad_rate", "WoE", "IV_contrib"]].round(3)
```

<div style="background:linear-gradient(90deg,#7C3AED 0%,#A66CFF 100%);border-radius:10px;padding:14px 22px;margin:26px 0 10px 0;"><span style="color:#fff;font-size:26px;font-weight:900;letter-spacing:.5px;"><span style="opacity:.6;">8.</span>&nbsp;&nbsp;PRÉ-PROCESSAMENTO COMO PIPELINE</span></div>

<div style="border-left:5px solid #7C3AED;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Por que pipeline</b><br><span style="color:#2D2A32;">Encapsular imputação, one-hot e escalonamento dentro de um <code>Pipeline</code> garante que a validação cruzada re-aprenda as transformações a cada fold, sem vazamento, e que o mesmo objeto sirva treino e produção. É a diferença entre um notebook e algo implantável.</span></div>

```python
# Removemos Sex por ser variável sensível e de IV desprezível
model_cats = [c for c in cat_cols if c != "Sex"]

preprocess = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore", drop="first"), model_cats),
    ("num", StandardScaler(), num_cols),
])
print("Categóricas no modelo:", model_cats)
print("Numéricas no modelo  :", num_cols)
```

<div style="background:linear-gradient(90deg,#7C3AED 0%,#A66CFF 100%);border-radius:10px;padding:14px 22px;margin:26px 0 10px 0;"><span style="color:#fff;font-size:26px;font-weight:900;letter-spacing:.5px;"><span style="opacity:.6;">9.</span>&nbsp;&nbsp;MODELAGEM: BASELINE E CHALLENGER</span></div>

<p style="color:#2D2A32;line-height:1.6;">
Dois modelos, dois papéis. A <b>regressão logística</b> é o baseline interpretável que uma área de
risco consegue defender diante de um regulador. O <b>XGBoost</b> é o challenger que mede quanto de
estrutura não-linear existe. Ambos recebem correção de desbalanceamento
(<code>class_weight</code> / <code>scale_pos_weight</code>) para não ignorar a classe minoritária,
e ambos são avaliados por <b>validação cruzada estratificada</b> antes de qualquer olhar no teste.</p>

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
scale_pos = (y_train == 0).sum() / (y_train == 1).sum()   # ~2.33

logit = Pipeline([
    ("prep", preprocess),
    ("clf", LogisticRegression(max_iter=1000, class_weight="balanced",
                               random_state=RANDOM_STATE)),
])
xgb = Pipeline([
    ("prep", preprocess),
    ("clf", XGBClassifier(n_estimators=250, max_depth=3, learning_rate=0.05,
                          subsample=0.8, colsample_bytree=0.8,
                          scale_pos_weight=scale_pos, eval_metric="auc",
                          random_state=RANDOM_STATE)),
])

results = []
for name, model in [("Regressão Logística", logit), ("XGBoost", xgb)]:
    auc_cv = cross_val_score(model, X_train, y_train, cv=cv, scoring="roc_auc")
    model.fit(X_train, y_train)
    p_test = model.predict_proba(X_test)[:, 1]
    results.append({
        "Modelo": name,
        "AUC (CV)": f"{auc_cv.mean():.3f} ± {auc_cv.std():.3f}",
        "AUC (teste)": roc_auc_score(y_test, p_test),
        "KS (teste)":  ks_statistic(y_test, p_test),
        "Gini (teste)": gini(y_test, p_test),
    })
res_df = pd.DataFrame(results).set_index("Modelo")
res_df.round(3)
```

<div style="border-left:5px solid #7C3AED;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Leitura</b><br><span style="color:#2D2A32;">O XGBoost ganha alguns pontos de AUC/Gini, mas a diferença é pequena e o intervalo de CV se sobrepõe. Em crédito, essa margem raramente compensa a perda de interpretabilidade e a complexidade de governança. Seguimos com a <b>logística como modelo campeão</b> e usamos o XGBoost como teto de referência.</span></div>

<div style="background:linear-gradient(90deg,#7C3AED 0%,#A66CFF 100%);border-radius:10px;padding:14px 22px;margin:26px 0 10px 0;"><span style="color:#fff;font-size:26px;font-weight:900;letter-spacing:.5px;"><span style="opacity:.6;">10.</span>&nbsp;&nbsp;AVALIAÇÃO NO CONJUNTO DE TESTE</span></div>

<h3 style="color:#7C3AED;">10.1 &nbsp;ROC e curva KS</h3>

```python
p_logit = logit.predict_proba(X_test)[:, 1]
p_xgb   = xgb.predict_proba(X_test)[:, 1]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5.2))

# --- ROC ---
for p, name, c in [(p_logit, "Logística", PURPLE), (p_xgb, "XGBoost", ORANGE)]:
    fpr, tpr, _ = roc_curve(y_test, p)
    ax1.plot(fpr, tpr, color=c, linewidth=2.5,
             label=f"{name} (AUC {roc_auc_score(y_test, p):.3f})")
ax1.plot([0, 1], [0, 1], "--", color="#b8b0cc")
ax1.set_xlabel("Falso positivo"); ax1.set_ylabel("Verdadeiro positivo")
ax1.set_title("Curva ROC"); ax1.legend(frameon=False, loc="lower right")

# --- KS ---
fpr, tpr, thr = roc_curve(y_test, p_logit)
ks_idx = np.argmax(tpr - fpr)
ax2.plot(thr, tpr, color=PURPLE, linewidth=2.2, label="% inadimplentes captados")
ax2.plot(thr, fpr, color=ORANGE, linewidth=2.2, label="% adimplentes captados")
ax2.vlines(thr[ks_idx], fpr[ks_idx], tpr[ks_idx], color=INK, linewidth=2)
ax2.text(thr[ks_idx]+.02, (tpr[ks_idx]+fpr[ks_idx])/2,
         f"KS = {tpr[ks_idx]-fpr[ks_idx]:.3f}", fontweight="bold", color=INK)
ax2.set_xlim(0, 1); ax2.set_xlabel("Ponto de corte (score)"); ax2.set_ylabel("Proporção acumulada")
ax2.set_title("Estatística KS — Logística"); ax2.legend(frameon=False, loc="center right")
plt.tight_layout(); plt.show()
```

<h3 style="color:#7C3AED;">10.2 &nbsp;Matriz de confusão e escolha do corte</h3>

<div style="border-left:5px solid #EA580C;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">O ponto que o negócio decide</b><br><span style="color:#2D2A32;">O corte padrão de 0,50 não é sagrado. Como reprovar um bom pagador custa margem e aprovar um mau custa o principal, o corte deve refletir esse custo. Abaixo comparamos 0,50 com um corte mais conservador que prioriza <b>capturar inadimplentes</b> (recall da classe 1), aceitando reprovar mais bons pagadores em troca.</span></div>

```python
def confusion_at(threshold, ax):
    pred = (p_logit >= threshold).astype(int)
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap=SEQ, cbar=False, ax=ax,
                annot_kws={"size": 15, "weight": "bold"},
                xticklabels=["Prev. bom", "Prev. mau"],
                yticklabels=["Real bom", "Real mau"], linewidths=2, linecolor="white")
    rec = cm[1, 1] / cm[1].sum()
    prec = cm[1, 1] / cm[:, 1].sum() if cm[:, 1].sum() else 0
    ax.set_title(f"Corte {threshold:.2f}\nRecall default {rec:.0%} · Precisão {prec:.0%}")

fig, axes = plt.subplots(1, 2, figsize=(12, 4.6))
confusion_at(0.50, axes[0])
confusion_at(0.60, axes[1])
plt.tight_layout(); plt.show()

print(classification_report(y_test, (p_logit >= 0.50).astype(int),
      target_names=["Adimplente", "Inadimplente"], digits=3))
```

<div style="border-left:5px solid #7C3AED;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Leitura</b><br><span style="color:#2D2A32;">Com balanceamento de classes, o modelo captura cerca de <b>três em cada quatro inadimplentes</b> no corte de 0,50 — bem acima dos ~35% do modelo original sem correção. O preço é maior taxa de falsos positivos (bons pagadores reprovados). Onde exatamente cravar o corte é decisão de política de crédito, calibrada pela matriz de custo real da instituição, não pelo default do sklearn.</span></div>

<div style="background:linear-gradient(90deg,#7C3AED 0%,#A66CFF 100%);border-radius:10px;padding:14px 22px;margin:26px 0 10px 0;"><span style="color:#fff;font-size:26px;font-weight:900;letter-spacing:.5px;"><span style="opacity:.6;">11.</span>&nbsp;&nbsp;INTERPRETAÇÃO E SCORECARD (PDO)</span></div>

```python
feat_names = logit.named_steps["prep"].get_feature_names_out()
coefs = logit.named_steps["clf"].coef_[0]
coef_tbl = (pd.DataFrame({"variavel": feat_names, "coeficiente": coefs})
            .assign(odds_ratio=lambda d: np.exp(d["coeficiente"]))
            .sort_values("coeficiente", key=abs, ascending=False))

fig, ax = plt.subplots(figsize=(9, 6))
top = coef_tbl.head(12).iloc[::-1]
colors = [ORANGE if c > 0 else PURPLE for c in top["coeficiente"]]
ax.barh(top["variavel"].str.replace("cat__|num__", "", regex=True),
        top["coeficiente"], color=colors, edgecolor="white")
ax.axvline(0, color=INK, linewidth=1)
ax.set_title("Coeficientes logísticos — laranja aumenta o risco, roxo reduz")
sns.despine(); plt.tight_layout(); plt.show()
coef_tbl.head(10).round(3)
```

<p style="color:#2D2A32;line-height:1.6;">
Um scorecard traduz a probabilidade em <b>pontos</b>, calibrados por dois parâmetros da indústria:
o <b>PDO</b> (points to double the odds — de quantos pontos as chances dobram) e um score de
referência. A fórmula usa <code>factor = PDO / ln(2)</code> e <code>offset = score − factor·ln(odds)</code>,
convertendo o log-odds do modelo numa escala legível de negócio.</p>

```python
def to_scorecard(prob, target_score=600, target_odds=20, pdo=20):
    factor = pdo / np.log(2)
    offset = target_score - factor * np.log(target_odds)
    odds = (1 - prob) / prob            # odds de ser bom pagador
    return offset + factor * np.log(odds)

scores = to_scorecard(p_logit)
fig, ax = plt.subplots(figsize=(9, 4.6))
for k, g in pd.DataFrame({"score": scores, "y": y_test.values}).groupby("y"):
    sns.kdeplot(g["score"], ax=ax, fill=True, alpha=.4, linewidth=2,
                color=PALETTE[k], label=("Inadimplente" if k else "Adimplente"))
ax.set_title("Distribuição de score de crédito (PDO = 20, ref. 600)")
ax.set_xlabel("Score"); ax.legend(frameon=False)
sns.despine(); plt.tight_layout(); plt.show()
print(f"Score médio — adimplentes: {scores[y_test.values==0].mean():.0f}"
      f"  |  inadimplentes: {scores[y_test.values==1].mean():.0f}")
```

<div style="border-left:5px solid #7C3AED;background:#faf7ff;border-radius:0 8px 8px 0;padding:12px 18px;margin:14px 0;"><b style="color:#7C3AED;">Leitura</b><br><span style="color:#2D2A32;">As duas distribuições de score se separam: bons pagadores concentram-se em faixas mais altas. Essa é a saída que uma mesa de crédito consome — um número na escala de sempre, com um ponto de corte de política sobreposto, não uma probabilidade abstrata.</span></div>
